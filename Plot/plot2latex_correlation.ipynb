```python
import pickle
import pandas as pd
```

```python

data_final_result_simplify_dsml_wkd_path = '../model_dml/theta/theta_DSML_2_3_4_16_17_18_19all_simplify.pickle'
data_final_result_simplify_dml_path = '../model_dml/theta/theta_DML_2_3_4_16_17_18_19all_simplify.pickle'
data_final_result_simplify_lr_path = '../model_dml/theta/theta_LR_2_3_4_16_17_18_19_simplify.pickle'

data_final_result_simplify_dsml_wke_path = '../model_dml/theta/theta_DSML_0_6_16_17_18_19all_simplify.pickle'
```

```python
with open(data_final_result_simplify_dsml_wkd_path, 'rb') as f:
    data_final_result_simplify_dsml_wkd_df = pickle.load(f)
    data_final_result_simplify_dsml_wkd_latex_df =  data_final_result_simplify_dsml_wkd_df[['region_id','results_params','p_value']]
print(data_final_result_simplify_dsml_wkd_latex_df)
print(data_final_result_simplify_dsml_wkd_latex_df["results_params"].mean())   
    
```

```python
with open(data_final_result_simplify_dsml_wke_path, 'rb') as f:
    data_final_result_simplify_dsml_wke_df = pickle.load(f)
    data_final_result_simplify_dsml_wke_latex_df =  data_final_result_simplify_dsml_wke_df[['region_id','results_params','p_value']]
print(data_final_result_simplify_dsml_wke_latex_df)
print(data_final_result_simplify_dsml_wke_latex_df["results_params"].mean())  
```

```python
with open(data_final_result_simplify_dml_path, 'rb') as f:
    data_final_result_simplify_dml_df = pickle.load(f)
    data_final_result_simplify_dml_latex_df =  data_final_result_simplify_dml_df[['region_id','results_params','p_value']]
print(data_final_result_simplify_dml_latex_df)
print(data_final_result_simplify_dml_latex_df["results_params"].mean()) 
```

```python

with open(data_final_result_simplify_lr_path, 'rb') as f:
    data_final_result_simplify_lr_df = pickle.load(f)
    data_final_result_simplify_lr_latex_df =  data_final_result_simplify_lr_df[['region_id','results_params','p_value']]
print(data_final_result_simplify_lr_latex_df)
print(data_final_result_simplify_lr_latex_df["results_params"].mean()) 
```

```python
df2latex_df =  pd.merge(pd.merge(data_final_result_simplify_dsml_wkd_latex_df,data_final_result_simplify_dml_latex_df,on='region_id'),data_final_result_simplify_lr_latex_df,on='region_id')
df2latex_df = df2latex_df.round(4)
df2latex_df
```

```python
def significant_indicator(x):
    x=float(x)
    print(x)
    if x <= 0.001:
        significant_value = '***'
    elif x <= 0.01 and x  > 0.001:
        significant_value = '**'
    elif x <= 0.05 and x > 0.01:
        significant_value = '*'
    else:
        significant_value = ''
    
    return str(x)+significant_value 
```

```python
df2latex_df['p_value_y'] = df2latex_df['p_value_y'].apply(lambda x: significant_indicator(x))


df2latex_df['p_value_x'] = df2latex_df['p_value_x'].apply(lambda x: significant_indicator(x))
df2latex_df['p_value'] = df2latex_df['p_value'].apply(lambda x: significant_indicator(x))

df2latex_df
```

```python
print(df2latex_df.to_latex(index=False))
```

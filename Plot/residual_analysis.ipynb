```python
import pickle
import matplotlib.pyplot as plt

import matplotlib.gridspec as gridspec
import seaborn as sns; sns.set()
import SeabornFig2Grid as sfg

plt.style.use('seaborn-poster')
```

```python

# ids = [4, 12, 13, 24, 43, 45, 48, 50, 68, 75, 79, 87, 88, 90, 100, 107, 113, 114, 125, 137, 140, 141, 142, 143, 144,148, 151, 158, 161, 162, 163, 164, 170, 186, 209, 211, 224, 229, 230, 231, 232, 233, 234, 236, 237, 238, 239,246, 249, 261, 262, 263]
ids = [100, 186, 236, 263]

id0 = ids[0]
id1 = ids[1]
id2 = ids[2]
id3 = ids[3]

# Saturday, Sunday
# target_weekday = ['0','6']
# Tues, Wed, Thurs
target_weekday = ['2', '3', '4']
identify_target_weekday = '_'.join(target_weekday)

target_hour = ['16', '17', '18', '19']
identify_target_hour = '_'.join(target_hour)

# model = 'DML'
model = 'DSML'
# model = 'LR'

tag = 'all'

data_final_result_path = '../model_dml/theta/' + 'theta_' + model + '_' + identify_target_weekday + '_' + identify_target_hour + tag + '.pickle'
```

```python
def plot_fig(data_final_result_path,ids,model):
    
    with open(data_final_result_path, 'rb') as f:
        result = pickle.load(f)

#     fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

    
    # subfig0
    Y_res_total = result[id0]['Y_res_total']
    T_res_total = result[id0]['T_res_total']

    # Figures Path
    Y_res_fig_path = './visual_residual/' + str(id0) + 'Y_res_' + model+ '_' + identify_target_weekday + '.pdf'
    T_res_fig_path = './visual_residual/' + str(id0) + 'T_res' + model+ '_' + identify_target_weekday + '.pdf'
    Y_T_res_fig_path = './visual_residual/' + str(
        id0) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '.pdf'
    seaborn_fig_path = './visual_residual/' + str(
        id0) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '_seaborn.pdf'
    
    g0 = sns.jointplot(x=T_res_total, y=Y_res_total,
                              kind="reg", truncate=False,
                              color="m", height=7, scatter_kws = {'s': 2, 'linewidth': 1},
                              # xlim=[-20,None],
                              joint_kws = {'line_kws': {'linewidth': 0.5}}, ylim = (-8,8))

    regline = g0.ax_joint.get_lines()[0]
    regline.set_color('black')
    regline.set_zorder(5)
#     g0.set_axis_labels('', 'Residual of speed (miles/hour)')
#     g0.set_axis_labels('(a) Region' + ' '+ str(id2), 'Residual of speed (miles/hour)')
    g0.set_axis_labels('(c) Region' + ' '+ str(id2), chr(958) + ' (miles/hour)')

    plt.suptitle("Your title here")
    
     # subfig1
    Y_res_total = result[id1]['Y_res_total']
    T_res_total = result[id1]['T_res_total']

    # Figures Path
    Y_res_fig_path = './visual_residual/' + str(id1) + 'Y_res_' + model+ '_' + identify_target_weekday + '.pdf'
    T_res_fig_path = './visual_residual/' + str(id1) + 'T_res' + model+ '_' + identify_target_weekday + '.pdf'
    Y_T_res_fig_path = './visual_residual/' + str(
        id1) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '.pdf'
    seaborn_fig_path = './visual_residual/' + str(
        id1) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '_seaborn.pdf'
    
    g1 = sns.jointplot(x=T_res_total, y=Y_res_total,
                              kind="reg", truncate=False,
                              color="m", height=7, scatter_kws = {'s': 2, 'linewidth': 1},
                              # xlim=[-20,None],
                              joint_kws = {'line_kws': {'linewidth': 0.5}},ylim = (-8,8))

    regline = g1.ax_joint.get_lines()[0]
    regline.set_color('black')
    regline.set_zorder(5)
#     g1.set_axis_labels('Residual of PUDO number (times/5 mins)', '')
    g1.set_axis_labels('(d) Region' + ' '+ str(id3), '')

  
     # subfig2
    Y_res_total = result[id2]['Y_res_total']
    T_res_total = result[id2]['T_res_total']

    # Figures Path
    Y_res_fig_path = './visual_residual/' + str(id2) + 'Y_res_' + model+ '_' + identify_target_weekday + '.pdf'
    T_res_fig_path = './visual_residual/' + str(id2) + 'T_res' + model+ '_' + identify_target_weekday + '.pdf'
    Y_T_res_fig_path = './visual_residual/' + str(
        id2) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '.pdf'
    seaborn_fig_path = './visual_residual/' + str(
        id2) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '_seaborn.pdf'
    
    g2 = sns.jointplot(x=T_res_total, y=Y_res_total,
                              kind="reg", truncate=False,
                              color="m", height=7, scatter_kws = {'s': 2, 'linewidth': 1},
                              # xlim=[-20,None],
                              joint_kws = {'line_kws': {'linewidth': 0.5}},ylim = (-8,8))

    regline = g2.ax_joint.get_lines()[0]
    regline.set_color('black')
    regline.set_zorder(5)
#     g2.set_axis_labels('Residual of PUDO number (times/5 mins)', 'Residual of speed (miles/hour)')
#     g2.set_axis_labels(, '')
    g2.set_axis_labels(chr(949)+' (times/5 mins)', chr(958) + ' (miles/hour)')
    
#     g2.fig.suptitle('(c) Region' + ' '+ str(id2))
#     g2.ax_joint.collections[0].set_alpha(0)
#     g2.fig.tight_layout()
#     g2.fig.subplots_adjust(top=0.15)


     # subfig3
    Y_res_total = result[id3]['Y_res_total']
    T_res_total = result[id3]['T_res_total']

    # Figures Path
    Y_res_fig_path = './visual_residual/' + str(id3) + 'Y_res_' + model+ '_' + identify_target_weekday + '.pdf'
    T_res_fig_path = './visual_residual/' + str(id3) + 'T_res' + model+ '_' + identify_target_weekday + '.pdf'
    Y_T_res_fig_path = './visual_residual/' + str(
        id3) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '.pdf'
    seaborn_fig_path = './visual_residual/' + str(
        id3) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '_seaborn.pdf'
    
    g3 = sns.jointplot(x=T_res_total, y=Y_res_total,
                              kind="reg", truncate=False,
                              color="m", height=7, scatter_kws = {'s': 2, 'linewidth': 1},
                              # xlim=[-20,None],
                              joint_kws = {'line_kws': {'linewidth': 0.5}},ylim = (-8,8))

    regline = g3.ax_joint.get_lines()[0]
    regline.set_color('black')
    regline.set_zorder(5)
    g3.set_axis_labels(chr(949)+' (times/5 mins)',  '')
   
    
#     g3.fig.suptitle('(d) Region' + ' '+ str(id3))
#     g2.ax_joint.collections[0].set_alpha(0)
#     g2.fig.tight_layout()
#     g3.fig.subplots_adjust(top=0.15)

    fig = plt.figure(figsize=(16,16))
    
#     fig.text(0.5, 0, 'Times of vehicles controlled', ha='center', va='center')
#     fig.text(0.06, 0.5, 'Improvement in total travel time by rerouting', ha='center', va='center', rotation='vertical')
    
    gs = gridspec.GridSpec(2, 2)

    mg0 = sfg.SeabornFig2Grid(g0, fig, gs[0,0])
    mg1 = sfg.SeabornFig2Grid(g1, fig, gs[0,1])
    mg2 = sfg.SeabornFig2Grid(g2, fig, gs[1,0])
    mg3 = sfg.SeabornFig2Grid(g3, fig, gs[1,1])

    gs.tight_layout(fig)
#     plt.subplots_adjust(wspace =0.05, hspace =1)
#     fig.text(0.5,-0.05, 'Residual of PUDO number (times/5 mins)', ha='center', va='center', fontsize=20)

#     fig.text(-1, 0.5, 'Residual of speed (miles/hour)', ha='center', va='center', rotation='vertical', fontsize=20)

    seaborn_fig_path = './visual_residual/' + str(
        id0) + str(id1) + str(id2) + str(id3) + 'Y_T_res_' + model+ '_' + identify_target_weekday + '_seaborn.eps'
    plt.savefig(seaborn_fig_path, bbox_inches="tight")
    plt.show()
    
```

```python
import pickle
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
data_final_result_simplify_path_pref = '../model_dml/theta/' 

theta_DML_2_3_4_16_17_18_19all_simplify_path = data_final_result_simplify_path_pref + 'theta_DML_2_3_4_16_17_18_19all_simplify.pickle'

theta_LR_2_3_4_16_17_18_19_simplify_path = data_final_result_simplify_path_pref + 'theta_LR_2_3_4_16_17_18_19_simplify.pickle'

theta_DSML_2_3_4_16_17_18_19_simplify_path = data_final_result_simplify_path_pref + 'theta_DSML_2_3_4_16_17_18_19_simplify.pickle'

theta_DSML_0_6_16_17_18_19all_simplify_path = data_final_result_simplify_path_pref + 'theta_DSML_0_6_16_17_18_19all_simplify.pickle'

theta_DSML_2_3_4_16_17_18_19_simplify_path_fig = './visual_theta_density/theta_DSML_2_3_4_0_6_16_17_18_19_simplify_path_fig.eps'

theta_DMLLR_2_3_4_16_17_18_19_simplify_path_fig = './visual_theta_density/theta_DML_LR_2_3_4_16_17_18_19_simplify_path_fig.eps'
```

```python

def DSML_wkd_wke(theta_DSML_2_3_4_16_17_18_19_simplify_path, theta_DSML_0_6_16_17_18_19all_simplify_path):
    
    with open(theta_DSML_2_3_4_16_17_18_19_simplify_path, 'rb') as f:
        theta_DSML_2_3_4_16_17_18_19_simplify_df = pickle.load(f)
    
    with open(theta_DSML_0_6_16_17_18_19all_simplify_path, 'rb') as f:
        theta_DSML_0_6_16_17_18_19all_simplify_df = pickle.load(f)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5), sharey=True)

#     g0 = sns.histplot(data=theta_DSML_2_3_4_16_17_18_19_simplify_df, x="results_params", kde=True, bins=10, 
#             line_kws={'lw': 3, 'c':'black'},  ax=ax1)
    
  
    theta_DSML_2_3_4_16_17_18_19_simplify_list = theta_DSML_2_3_4_16_17_18_19_simplify_df['results_params'].values.tolist()
#     print(theta_DSML_2_3_4_16_17_18_19_simplify_list)
    g0 = sns.distplot(theta_DSML_2_3_4_16_17_18_19_simplify_list,
                  kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                  hist_kws={ "linewidth": 3, 
                            "alpha": 0.8, "color": "m"}, ax=ax1,  bins=10)
    g0.set(xlabel= chr(952), ylabel='')
    
    
#     g0.grid()

    theta_DSML_0_6_16_17_18_19_simplify_list = theta_DSML_0_6_16_17_18_19all_simplify_df['results_params'].values.tolist()
#     print(theta_DSML_0_6_16_17_18_19_simplify_list)
    g1 = sns.distplot(theta_DSML_0_6_16_17_18_19_simplify_list,
                  kde_kws={"color": "k", "lw": 3, "label": "KDE"},
                  hist_kws={ "linewidth": 3, 
                            "alpha": 0.8, "color": "m"}, ax=ax2,  bins=10)

    g1.set(xlabel=chr(952), ylabel='')
#     g1.fig.suptitle("(b) on weekends")
#     g1.grid()

#     fig.text(0.5, -0.06, 'Times of vehicles controlled', ha='center', va='center', fontsize=20)

    fig.text(0.075, 0.5, 'Density', ha='center', va='center', rotation='vertical', fontsize=20)
    
    plt.subplots_adjust(wspace =0.15, hspace =0.05)
    plt.savefig(theta_DSML_2_3_4_16_17_18_19_simplify_path_fig, bbox_inches="tight")
    plt.show()
    
```

```python
DSML_wkd_wke(theta_DSML_2_3_4_16_17_18_19_simplify_path, theta_DSML_0_6_16_17_18_19all_simplify_path)
```

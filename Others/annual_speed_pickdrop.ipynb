```python
data_for_training_wkd_path = './model_dml/training_data/training_data_dmsl_XWXMW_more_parameters2_3_4_16_17_18_19.pickle'
data_for_training_wke_path = './model_dml/training_data/training_datatraining_data_dmsl_XWXMW_more_parameters0_6_16_17_18_19.pickle'

input_region_geo_Manhattan_path = './data_set/NYC_geo/NYC_Taxi_Zones_Manhattan.geojson'
center_point_region_df_path = './Shortest_path/center_point_region_df.pickle'
ids = [4, 12, 13, 24, 43, 45, 48, 50, 68, 75, 79, 87, 88, 90, 100, 107, 113, 114, 125, 137, 140, 141, 142, 143, 144,148, 151, 158, 161, 162, 163, 164, 170, 186, 209, 211, 224, 229, 230, 231, 232, 233, 234, 236, 237, 238, 239,246, 249, 261, 262, 263]
center_point_region_df, region_polygon_gpd = get_center_point(input_region_geo_Manhattan_path, center_point_region_df_path, ids)
```

```python
theta_map_speed_wkd_path = './Visualization_plot/theta_map_speed_wkd.html'
theta_map_speed_wkd_html_path = './Visualization_plot/theta_map_speed_wkd_html.html'

theta_map_pickdrop_wkd_path = './Visualization_plot/theta_map_pickdrop_wkd.html'
theta_map_pickdrop_wkd_html_path = './Visualization_plot/theta_map_pickdrop_wkd_html.html'


theta_map_speed_wke_path = './Visualization_plot/theta_map_speed_wke.html'
theta_map_speed_wke_html_path = './Visualization_plot/theta_map_speed_wke_html.html'

theta_map_pickdrop_wke_path = './Visualization_plot/theta_map_pickdrop_wke.html'
theta_map_pickdrop_wke_html_path = './Visualization_plot/theta_map_pickdrop_wke_html.html'
```

```python
def read_ave_annual(data_for_training_wkd_path, data_for_training_wke_path, ids):
    
    with open(data_for_training_wkd_path, 'rb') as f:
        training_data_for_evening_wkd_df = pickle.load(f)
    
    with open(data_for_training_wke_path, 'rb') as f:
        training_data_for_evening_wke_df = pickle.load(f)
    
    training_data_for_evening_plot_speed_wkd_df = training_data_for_evening_wkd_df.groupby('region_id')['relative_speed'].mean()
    training_data_for_evening_plot_speed_wkd_df = training_data_for_evening_plot_speed_wkd_df.reset_index()
    training_data_for_evening_plot_speed_wkd_df['region_id'] = training_data_for_evening_plot_speed_wkd_df['region_id'].apply(lambda x: int(x))
    training_data_for_evening_plot_speed_wkd_df = training_data_for_evening_plot_speed_wkd_df[training_data_for_evening_plot_speed_wkd_df["region_id"].isin(ids)]
    
    training_data_for_evening_plot_pickdrop_wkd_df = training_data_for_evening_wkd_df.groupby('region_id')['total_number'].mean()
    training_data_for_evening_plot_pickdrop_wkd_df = training_data_for_evening_plot_pickdrop_wkd_df.reset_index()
    training_data_for_evening_plot_pickdrop_wkd_df['region_id'] = training_data_for_evening_plot_pickdrop_wkd_df['region_id'].apply(lambda x: int(x))
    training_data_for_evening_plot_pickdrop_wkd_df = training_data_for_evening_plot_pickdrop_wkd_df[training_data_for_evening_plot_pickdrop_wkd_df["region_id"].isin(ids)]
    
     
    training_data_for_evening_plot_speed_wke_df = training_data_for_evening_wke_df.groupby('region_id')['relative_speed'].mean()
    training_data_for_evening_plot_speed_wke_df = training_data_for_evening_plot_speed_wke_df.reset_index()
    training_data_for_evening_plot_speed_wke_df['region_id'] = training_data_for_evening_plot_speed_wke_df['region_id'].apply(lambda x: int(x))
    training_data_for_evening_plot_speed_wke_df = training_data_for_evening_plot_speed_wke_df[training_data_for_evening_plot_speed_wke_df["region_id"].isin(ids)]
    
    training_data_for_evening_plot_pickdrop_wke_df = training_data_for_evening_wke_df.groupby('region_id')['total_number'].mean()
    training_data_for_evening_plot_pickdrop_wke_df = training_data_for_evening_plot_pickdrop_wke_df.reset_index()
    training_data_for_evening_plot_pickdrop_wke_df['region_id'] = training_data_for_evening_plot_pickdrop_wke_df['region_id'].apply(lambda x: int(x))
    training_data_for_evening_plot_pickdrop_wke_df = training_data_for_evening_plot_pickdrop_wke_df[training_data_for_evening_plot_pickdrop_wke_df["region_id"].isin(ids)]
    
    return training_data_for_evening_plot_speed_wkd_df, training_data_for_evening_plot_pickdrop_wkd_df,training_data_for_evening_plot_speed_wke_df,training_data_for_evening_plot_pickdrop_wke_df
```

```python
def plot_map_speed_pickdrop(training_data_for_evening_plot_speed_wkd_df, training_data_for_evening_plot_pickdrop_wkd_df,training_data_for_evening_plot_speed_wke_df,training_data_for_evening_plot_pickdrop_wke_df, region_polygon_gpd):
    
    bins_speed = [i for i in range(7, 33, 5)]
    bin_pickdrop = [i for i in range(0, 160, 31)]

    m = folium.Map([40.730610, -73.935242], zoom_start=12, tiles=None)
    n = folium.Map([40.730610, -73.935242], zoom_start=12, tiles=None)
    
    p = folium.Map([40.730610, -73.935242], zoom_start=12, tiles=None)
    q = folium.Map([40.730610, -73.935242], zoom_start=12, tiles=None)
    
    #tiles = ['stamenwatercolor', 'cartodbpositron', 'openstreetmap', 'stamenterrain']
    
    training_data_for_evening_plot_speed_wkd_df = training_data_for_evening_plot_speed_wkd_df.reset_index()
    training_data_for_evening_plot_pickdrop_wkd_df = training_data_for_evening_plot_pickdrop_wkd_df.reset_index()
    training_data_for_evening_plot_speed_wke_df = training_data_for_evening_plot_speed_wke_df.reset_index()
    training_data_for_evening_plot_pickdrop_wke_df = training_data_for_evening_plot_pickdrop_wke_df.reset_index()
    region_polygon_gpd = region_polygon_gpd.reset_index()
    
    region_polygon_gpd['relative_speed_wkd'] = training_data_for_evening_plot_speed_wkd_df['relative_speed']
    region_polygon_gpd['relative_speed_wke'] = training_data_for_evening_plot_speed_wke_df['relative_speed']
    region_polygon_gpd['total_number_wkd'] = training_data_for_evening_plot_pickdrop_wkd_df['total_number']
    region_polygon_gpd['total_number_wke'] = training_data_for_evening_plot_pickdrop_wke_df['total_number']
    
    add_choropleth_speed_wkd = folium.Choropleth(geo_data=region_polygon_gpd, data=training_data_for_evening_plot_speed_wkd_df,
                                       columns=['region_id', 'relative_speed'],
                                       key_on='feature.properties.region_id', 
                                       fill_color='YlGn',
                                       fill_opacity=0.7, line_opacity=0.2,
                                       legend_name='Average of Speed in Regions on Weekdays',
                                       highlight=True,
                                       bins=bins_speed
                                       ).add_to(m)
    
    add_choropleth_speed_wkd.geojson.add_child(folium.features.GeoJsonTooltip(['region_id','relative_speed_wkd']))
    m.add_child(folium.ClickForMarker(popup="Waypoint"))
    m.add_child(folium.map.LayerControl())
    m.add_child(folium.LatLngPopup())
    m.add_child(folium.LayerControl())
    
    m.save(theta_map_speed_wkd_path)
    webbrowser.open(theta_map_speed_wkd_html_path)
    
    add_choropleth_speed_wke = folium.Choropleth(geo_data=region_polygon_gpd, data=training_data_for_evening_plot_speed_wke_df,
                                   columns=['region_id', 'relative_speed'],
                                   key_on='feature.properties.region_id', 
                                   fill_color='YlGn',
                                   fill_opacity=0.7, line_opacity=0.2,
                                   legend_name='Average of Speed in Regions on Weekends',
                                   highlight=True,
                                   bins=bins_speed
                                   ).add_to(p)
    
    add_choropleth_speed_wke.geojson.add_child(folium.features.GeoJsonTooltip(['region_id','relative_speed_wke']))
    p.add_child(folium.ClickForMarker(popup="Waypoint"))
    p.add_child(folium.map.LayerControl())
    p.add_child(folium.LatLngPopup())
    p.add_child(folium.LayerControl())
    
    p.save(theta_map_speed_wke_path)
    webbrowser.open(theta_map_speed_wke_html_path)
    
    
    add_choropleth_pickdrop_wkd = folium.Choropleth(geo_data=region_polygon_gpd, data=training_data_for_evening_plot_pickdrop_wkd_df,
                                   columns=['region_id', 'total_number'],
                                   key_on='feature.properties.region_id',          
                                   fill_color='Blues',
                                   fill_opacity=0.7, line_opacity=0.2,
                                   legend_name='Average of number of Pick-up/Drop-off in Regions on Weekdays',
                                   highlight=True,
                                   bins=bin_pickdrop
                                   ).add_to(n)
    
    add_choropleth_pickdrop_wkd.geojson.add_child(folium.features.GeoJsonTooltip(['region_id','total_number_wkd']))
    n.add_child(folium.ClickForMarker(popup="Waypoint"))
    n.add_child(folium.map.LayerControl())
    n.add_child(folium.LatLngPopup())
    n.add_child(folium.LayerControl())
    
    n.save(theta_map_pickdrop_wkd_path)
    webbrowser.open(theta_map_pickdrop_wkd_html_path)

    
    
    add_choropleth_pickdrop_wke = folium.Choropleth(geo_data=region_polygon_gpd, data=training_data_for_evening_plot_pickdrop_wke_df,
                               columns=['region_id', 'total_number'],
                               key_on='feature.properties.region_id',          
                               fill_color='Blues',
                               fill_opacity=0.7, line_opacity=0.2,
                               legend_name='Average of number of Pick-up/Drop-off in Regions on Weekends',
                               highlight=True,
                               bins=bin_pickdrop
                               ).add_to(q)
    
    add_choropleth_pickdrop_wke.geojson.add_child(folium.features.GeoJsonTooltip(['region_id','total_number_wke']))
    q.add_child(folium.ClickForMarker(popup="Waypoint"))
    q.add_child(folium.map.LayerControl())
    q.add_child(folium.LatLngPopup())
    q.add_child(folium.LayerControl())
    
    q.save(theta_map_pickdrop_wke_path)
    webbrowser.open(theta_map_pickdrop_wke_html_path)
    
    return m,n,p,q
```

```python
if __name__ == '__main__':
    
    training_data_for_evening_plot_speed_wkd_df, training_data_for_evening_plot_pickdrop_wkd_df,training_data_for_evening_plot_speed_wke_df,training_data_for_evening_plot_pickdrop_wke_df = read_ave_annual(data_for_training_wkd_path, data_for_training_wke_path, ids)
    
    m,n,p,q = plot_map_speed_pickdrop(training_data_for_evening_plot_speed_wkd_df, training_data_for_evening_plot_pickdrop_wkd_df,training_data_for_evening_plot_speed_wke_df,training_data_for_evening_plot_pickdrop_wke_df
, region_polygon_gpd)
    
    
    
```
